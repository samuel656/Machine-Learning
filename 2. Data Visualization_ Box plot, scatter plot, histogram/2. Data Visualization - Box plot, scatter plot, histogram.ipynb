```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['gender'].value_counts().plot(kind='bar', title='Gender of students')
plt.xticks(rotation=0)
plt.show()
```

```python
df['race/ethnicity'].value_counts().plot(kind='bar', title='Race/ethnicity of students')
plt.xticks(rotation=0)
plt.show()
```

```python
df['lunch'].value_counts().plot(kind='bar', title='Lunch status of students')
plt.xticks(rotation=0)
plt.show()
```

```python
df['test preparation course'].value_counts().plot(kind='bar', title='Test preparation course')
plt.xticks(rotation=0)
plt.show()
```

```python
df.boxplot()
```

```python
sns.distplot(df['math score'])
```

```python
corr = df.corr()
sns.heatmap(corr, annot=True, square=True)
plt.yticks(rotation=0)
plt.show()
```

```python
sns.scatterplot(x='math score', y='writing score', hue='gender', data=df)
```
